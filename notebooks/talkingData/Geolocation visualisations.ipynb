```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.basemap import Basemap
%matplotlib inline
```

```python
df_events = pd.read_csv("~/百度云同步盘/dev/Kaggle/TalkingData/events.csv", dtype={'device_id': np.str})
```

```python
df_events.head()
```

### Plot a bunch of different maps showing the locations of the events

```python
# Set up plot
df_events_sample = df_events.sample(n=100000)
plt.figure(1, figsize=(12, 6))

# Mercator of World
map1 = Basemap(projection='merc',
              llcrnrlat=-60,
              urcrnrlat=65,
              llcrnrlon=-180,
              urcrnrlon=180,
              lat_ts=0,
              resolution='c')

map1.fillcontinents(color='#191919',lake_color='#000000') # dark grey land, black lakes
map1.drawmapboundary(fill_color='#000000')                # black background
map1.drawcountries(linewidth=0.1, color='w')              # thin white line for country borders

# Plot the data
mxy = map1(df_events_sample["longitude"].tolist(), df_events_sample["latitude"].tolist())
map1.scatter(mxy[0], mxy[1], s=3, c="#1292db", lw=0, alpha=1, zorder=5)

plt.title("Global view of events")
plt.show()
```

Not surprisingly, most of the events are geo-located in China. We also see a few sporadic ones around the globe, most which look real (e.g., Austalia-Sydney/Melbourne/Perth). <br/><br/>
One interesting thing is there are a number of events at (lat,lon) = (0,0), and also around that area where there is not much land.

```python
df_at0 = df_events[(df_events["longitude"]==0) & (df_events["latitude"]==0)]
df_near0 = df_events[(df_events["longitude"] > -1)
                    & (df_events["longitude"] < 1)
                    & (df_events["latitude"] > -1)
                    & (df_events["latitude"] < 1)]

print("# events:", len(df_events))
print("# df_at0:", len(df_at0))
print("# df_near0", len(df_near0))
```

** print another mercator plot, but zooming in on China **

```python
# Sample it down to only the China region
lon_min, lon_max = 75, 135
lat_min, lat_max = 15, 55

idx_china = (df_events["longitude"] > lon_min) &\
            (df_events["longitude"] < lon_max) &\
            (df_events["latitude"] > lat_min) &\
            (df_events["latitude"] < lat_max)

df_events_china = df_events[idx_china].sample(n=100000)
```

```python
# Mercator of China
plt.figure(2, figsize=(12, 6))

map2 = Basemap(projection='merc',
             llcrnrlat=lat_min,
             urcrnrlat=lat_max,
             llcrnrlon=lon_min,
             urcrnrlon=lon_max,
             lat_ts=35,
             resolution='i')

map2.fillcontinents(color='#191919',lake_color='#000000') # dark grey land, black lakes
map2.drawmapboundary(fill_color='#000000')                # black background
map2.drawcountries(linewidth=0.1, color="w")              # thin white line for country borders

# Plot the data
mxy = map2(df_events_china["longitude"].tolist(), df_events_china["latitude"].tolist())
map2.scatter(mxy[0], mxy[1], s=5, c="#1292db", lw=0, alpha=0.05, zorder=5)

plt.title("China view of events")
plt.show()
```

This is more interesting to visualise, and roughly matches what you'd expect from the population density in China.

```python
# Sample it down to only the Beijing region
lon_min, lon_max = 116, 117
lat_min, lat_max = 39.75, 40.25

idx_beijing = (df_events["longitude"]>lon_min) &\
              (df_events["longitude"]<lon_max) &\
              (df_events["latitude"]>lat_min) &\
              (df_events["latitude"]<lat_max)

df_events_beijing = df_events[idx_beijing]

# Mercator of Beijing
plt.figure(3, figsize=(12,6))

map3 = Basemap(projection='merc',
             llcrnrlat=lat_min,
             urcrnrlat=lat_max,
             llcrnrlon=lon_min,
             urcrnrlon=lon_max,
             lat_ts=35,
             resolution='c')

map3.fillcontinents(color='#191919',lake_color='#000000') # dark grey land, black lakes
map3.drawmapboundary(fill_color='#000000')                # black background
map3.drawcountries(linewidth=0.1, color="w")              # thin white line for country borders

# Plot the data
mxy = map3(df_events_beijing["longitude"].tolist(), df_events_beijing["latitude"].tolist())
map3.scatter(mxy[0], mxy[1], s=5, c="#1292db", lw=0, alpha=0.1, zorder=5)

plt.title("Beijing view of events")
plt.show()
```


At this scale, you can actually see the finite resolution of the lat/lon values.


### Now we can try plotting the same for male/female, and show average age per grid

```python
# Load the train data and join the events
df_train = pd.read_csv("~/百度云同步盘/dev/Kaggle/TalkingData/gender_age_train.csv", dtype={'device_id': np.str})

df_plot = pd.merge(df_train, df_events_beijing, on='device_id', how='inner')

df_m = df_plot[df_plot["gender"]=="M"]
df_f = df_plot[df_plot["gender"]=="F"]
```

```python
# Male/Female plot
plt.figure(4, figsize=(12, 6))

plt.subplot(121)

map4a = Basemap(projection='merc',
             llcrnrlat=lat_min,
             urcrnrlat=lat_max,
             llcrnrlon=lon_min,
             urcrnrlon=lon_max,
             lat_ts=35,
             resolution='c')

map4a.fillcontinents(color='#191919',lake_color='#000000') # dark grey land, black lakes
map4a.drawmapboundary(fill_color='#000000')                # black background
map4a.drawcountries(linewidth=0.1, color="w")              # thin white line for country borders

mxy = map4a(df_m["longitude"].tolist(), df_m["latitude"].tolist())
map4a.scatter(mxy[0], mxy[1], s=5, c="#1292db", lw=0, alpha=0.1, zorder=5)
plt.title("Male events in Beijing")

plt.subplot(122)

map4b = Basemap(projection='merc',
             llcrnrlat=lat_min,
             urcrnrlat=lat_max,
             llcrnrlon=lon_min,
             urcrnrlon=lon_max,
             lat_ts=35,
             resolution='c')

map4b.fillcontinents(color='#191919',lake_color='#000000') # dark grey land, black lakes
map4b.drawmapboundary(fill_color='#000000')                # black background
map4b.drawcountries(linewidth=0.1, color="w")              # thin white line for country borders

mxy = map4b(df_f["longitude"].tolist(), df_f["latitude"].tolist())
map4b.scatter(mxy[0], mxy[1], s=5, c="#fd3096", lw=0, alpha=0.1, zorder=5)
plt.title("Female events in Beijing")

plt.show()
```

Clearly some different patterns emerging, though it's not immediately clear whether or not this is just because there are different amounts of data for M/F.

```python
# Make a pivot table showing average age per area of a grid, also store the counts
df_plot["lon_round"] = df_plot["longitude"].round(decimals=2)
df_plot["lat_round"] = df_plot["latitude"].round(decimals=2)

df_age = pd.pivot_table(df_plot,
                       values='age',
                       index='lon_round',
                       columns='lat_round',
                       aggfunc=np.mean)

df_cnt = pd.pivot_table(df_plot,
                       values='age',
                       index='lon_round',
                       columns='lat_round',
                       aggfunc="count")
```

```python
# Age plot
plt.figure(5, figsize=(12, 6))

# plot avg age per grid
plt.subplot(121)
map5a = Basemap(projection='merc',
               llcrnrlat=lat_min,
               urcrnrlat=lat_max,
               llcrnrlon=lon_min,
               urcrnrlon=lon_max,
               lat_ts=35,
               resolution='c')
# Construct a heatmap
lons = df_age.index.values
lats = df_age.columns.values
x, y = np.meshgrid(lons, lats)
px, py = map5a(x, y)
data_values = df_age.values
masked_data = np.ma.masked_invalid(data_values.T)
cmap = plt.cm.viridis
cmap.set_bad(color='#191919')
# Plot the heatmap
map5a.pcolormesh(px, py, masked_data, cmap=cmap, zorder=5)
map5a.colorbar().set_label("average age")
plt.title("Average age per grid area in Beijing")

# Plot count per grid
plt.subplot(122)
map5b = Basemap(projection='merc',
             llcrnrlat=lat_min,
             urcrnrlat=lat_max,
             llcrnrlon=lon_min,
             urcrnrlon=lon_max,
             lat_ts=35,
             resolution='c')      
# Construct a heatmap 
data_values = df_cnt.values
masked_data = np.ma.masked_invalid(data_values.T)
cmap = plt.cm.viridis
cmap.set_bad(color="#191919")
# Plot the heatmap
map5b.pcolormesh(px, py, masked_data, cmap=cmap, zorder=5)
map5b.colorbar().set_label("count")
plt.title("Event count per grid area in Beijing")

plt.show()
```
